```python
from PIL import Image, ImageFilter
from IPython import display


input_path = 'resized-images/3-red-striped-oval-0.90i-7deg.png'
input_path = 'resized-images/3-red-empty-oval-0.98i-347deg.png'
input_path = 'resized-images/3-red-solid-oval-0.80i-10deg.png'
display.display(display.Image(filename=input_path))

image = Image.open(input_path)
blurred = image.filter(ImageFilter.GaussianBlur(4))
blurred.save('/tmp/out.png')
display.display(display.Image(filename='/tmp/out.png'))
```
